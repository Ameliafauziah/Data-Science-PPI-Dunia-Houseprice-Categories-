# Descriptive Analysis

```python
import matplotlib.pyplot as plt
from scipy.stats import skew
import pandas as pd

for i in df:
    if pd.api.types.is_numeric_dtype(df[i]):
        nilai_skew = skew(df[i].dropna())

        # Interpretasi skew
        if nilai_skew > 0.5:
            interpretasi = "Distribusi condong ke kanan (positif skew)"
        elif nilai_skew < -0.5:
            interpretasi = "Distribusi condong ke kiri (negatif skew)"
        else:
            interpretasi = "Distribusi normal (simetris)"

        print(f"Skew dari kolom {i}: {nilai_skew:.2f} → {interpretasi}")
        plt.hist(df[i].dropna(), bins=30, edgecolor='black')
        plt.title(f'Distribusi {i}')
        plt.xlabel(i)
        plt.ylabel('Frekuensi')
        plt.show()
        print(" ")
        print("------------------------------------------------")
        print(" ")
```

## Penyebaran

```python
for i in df.select_dtypes(include='number').columns:
    print(f"Informasi Penyebaran untuk kolom: {i}")

    # Quantiles
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    print(f"Q1       : {Q1}")
    print(f"Q3       : {Q3}")

    # Dispersion Measures
    std_dev = df[i].std()
    variance = df[i].var()
    print(f"Standard Deviation (Std) : {std_dev}")
    print(f"Variance                  : {variance}")

    # IQR (Interquartile Range)
    IQR = Q3 - Q1
    print(f"IQR (Interquartile Range)  : {IQR}")

    # Range
    data_range = df[i].max() - df[i].min()
    print(f"Range                      : {data_range}")

    # Calculate outlier bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Detect outliers
    outliers = df[(df[i] < lower_bound) | (df[i] > upper_bound)]
    print(f"Outliers detected: {len(outliers)}")

    print("-" * 50)
```

## Uji Hipotesis

Case : Melakukan uji hipotesis menggunakan TTest untuk membuktikan apakah terdapat pengaruh antara terdapat atau tidaknya jumlah pemandangan air mempengaruhi harga rumah atau tidak.

Note : Disini, asumsi TTest bahwa data harus berdistribusi normal diabaikan karena kita belum mempelajari bagaimana caranya agar data berdistribusi normal.

```python
df.info()
```

### TTest

```python
df["waterfront"].unique()
```

```python
import scipy.stats as st

# Memisahkan data harga rumah berdasarkan keberadaan waterfront
with_waterfront = df[df['waterfront'] == 1]['price']
without_waterfront = df[df['waterfront'] == 0]['price']

# Menghitung mean masing-masing
mean_with = with_waterfront.mean()
mean_without = without_waterfront.mean()

# Melakukan uji t-test independen
t_stat, p_value = st.ttest_ind(with_waterfront, without_waterfront, equal_var=False)

# Menampilkan hasil
print(f"Mean price dengan waterfront: {mean_with}")
print(f"Mean price tanpa waterfront: {mean_without}")
print(f"P-value: {p_value}")

# Interpretasi
if p_value < 0.05:
    print("Terima H1, cukup bukti mean price berbeda signifikan")
else:
    print("Gagal tolak H0, tidak cukup bukti perbedaan signifikan pada mean price")
```
